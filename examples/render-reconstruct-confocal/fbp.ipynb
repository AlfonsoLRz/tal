```python
import tal

data = tal.io.read_capture('confocal-scene.hdf5')  # you'll need to generate this file using "tal render exhaustive scene"
tal.reconstruct.compensate_laser_cos_dsqr(data)
```

```python
volume_xyz = tal.reconstruct.get_volume_project_rw(data, depths=[1.0,])
```

```python
H_1_fbp = tal.reconstruct.fbp.solve(data, wl_mean=0.25, wl_sigma=0.25, volume_xyz=volume_xyz, camera_system=tal.enums.CameraSystem.DIRECT_LIGHT)
```

```python
tal.plot.amplitude_phase(H_1_fbp, title='Reconstruction')
```

```python
with tal.resources(cpu_processes=2, downscale=2):
    H_1_fbp = tal.reconstruct.fbp.solve(data, wl_mean=0.25, wl_sigma=0.25, volume_xyz=volume_xyz, camera_system=tal.enums.CameraSystem.DIRECT_LIGHT)
```

```python
tal.plot.amplitude_phase(H_1_fbp, title='Multiple threads')
```

```python
with tal.resources(cpu_processes=2, downscale=4):
    H_1_fbp = tal.reconstruct.fbp.solve(data, wl_mean=0.25, wl_sigma=0.25, volume_xyz=volume_xyz, camera_system=tal.enums.CameraSystem.DIRECT_LIGHT)
```

```python
tal.plot.amplitude_phase(H_1_fbp, title='Multiple threads')
```

```python
with tal.resources(cpu_processes=2, downscale=2):
    H_1_fbp = tal.reconstruct.fbp.solve(data, wl_mean=0.25, wl_sigma=0.25, volume_xyz=volume_xyz,
                                        projector_focus=[0, -0.5, 1],
                                        camera_system=tal.enums.CameraSystem.PROJECTOR_CAMERA_T0)
```

```python
tal.plot.amplitude_phase(H_1_fbp, title='Focus illumination')
```
